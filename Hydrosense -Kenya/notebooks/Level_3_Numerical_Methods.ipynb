# 2. Root Finding Problem

The objective is to find the irrigation amount I that allows soil moisture to reach a target value.

f(I)=predicted moisture-target moisture

The solution occurs when f(I)=0.

Root finding methods are numerical approaches used to determine the value of an unknown variable that satisfies a mathematical equation. In the HydroSense-Kenya irrigation system, these methods can be applied to estimate the amount of irrigation water required to achieve a target soil moisture level.


The following methods are investigated:

### Bisection Method

The bisection method is a reliable root-finding technique that repeatedly divides an interval into smaller sections until the solution is found. It is based on the principle that if a function changes sign within an interval, a root exists between those points.

This method is simple and stable, making it useful for scientific applications where reliability is more important than speed. However, it may require many iterations compared with faster methods.

### Newton-Raphson Method

The Newton-Raphson method improves solution speed by using the derivative of a function to estimate the direction toward the root. It uses the current estimate to calculate a better approximation in each iteration.

This method usually converges faster than bisection, but its performance depends on selecting a suitable starting point and having accurate derivative information.

### Secant Method

The secant method is similar to Newton-Raphson but avoids directly calculating derivatives. Instead, it estimates the slope using two previous values.

This provides a balance between simplicity and speed, making it useful when derivative calculations are difficult or unavailable.

ROOT FINDING PROBLEM

```python
def irrigation_function(x):

    return (
        0.12*x +
        0.35*x +
        2.4*x -
        20
    )


def irrigation_derivative(x):

    return 2.87
```

# Bisection Method

```python
root_bisection = bisection(
    irrigation_function,
    0,
    20
)


print("Bisection Root:")
print(root_bisection)
```

# Newton-Raphson Method

```python
root_newton = newton_raphson(
    irrigation_function,
    irrigation_derivative,
    5,
    1e-6,
    100
)


print("Newton-Raphson Result:")
print(root_newton)
```

# Secant Method

```python
#THE SECANT FUNCTION AS DEFINED IN THE src/numerical_methods.py FILE
root_secant = secant(
    irrigation_function,
    0,
    50
)
#the secant function is defined in the src/numerical_methods.py file, so we can call it directly here
print("Secant Result:")
print(root_secant)
```

## Numerical Convergence Analysis

Convergence analysis evaluates how quickly numerical methods approach the true
solution. The error decreases as the number of iterations increases.

Bisection has guaranteed but slower convergence because the interval is reduced
by half each iteration. Newton-Raphson usually converges faster because it uses
derivative information. The Secant method provides a balance by approximating
the derivative without requiring explicit calculation.

The convergence behaviour helps determine the most suitable method for solving
irrigation-related mathematical problems.

```python
bisection_error = [
    abs(item[2]) for item in root_bisection[3]
]


newton_error = [
    abs(item[2]) for item in root_newton[3]
]


secant_error = [
    abs(item[2]) for item in root_secant[3]
]



print("Bisection errors:", bisection_error)

print("Newton errors:", newton_error)

print("Secant errors:", secant_error)
```

```python
plt.figure(figsize=(8,5))


plt.semilogy(
    bisection_error,
    label="Bisection"
)


plt.semilogy(
    newton_error,
    label="Newton-Raphson"
)


plt.semilogy(
    secant_error,
    label="Secant"
)


plt.xlabel("Iteration")

plt.ylabel("Absolute Error")

plt.title(
    "Numerical Convergence Comparison of Root Finding Methods"
)


plt.legend()

plt.grid()


plt.show()
```

# 3. Numerical Differentiation

Finite differences estimate changes in soil moisture.

Numerical differentiation estimates the rate of change of a function using discrete data points. In agricultural systems, sensor measurements are collected at specific time intervals, making numerical differentiation useful for analysing changes in soil moisture, temperature, and water availability.

The following methods are investigated:

### Forward Difference

The forward difference method estimates the derivative using the current value and the next value. It is simple to implement but may have lower accuracy because it only considers future information.

### Backward Difference

The backward difference method uses the current value and the previous value to estimate change. It is useful when analysing historical sensor data where future values are unavailable.

### Central Difference

The central difference method uses values on both sides of a point, providing a more accurate approximation compared with forward and backward methods. It is commonly preferred when sufficient data points are available.

```python
x = np.arange(0,10,1)


y = np.sin(x)+2


h = 1


i = 5


forward_result = forward_difference(
    y,
    h,
    i
)


backward_result = backward_difference(
    y,
    h,
    i
)


central_result = central_difference(
    y,
    h,
    i
)



print("Forward Difference:", forward_result)

print("Backward Difference:", backward_result)

print("Central Difference:", central_result)
```

# 4. Numerical Integration

Integration estimates cumulative water deficit over time.

Numerical integration estimates the total accumulation of a quantity over a period of time. In HydroSense-Kenya, integration can be used to estimate cumulative evapotranspiration, total water loss, or irrigation requirements.

The following methods are investigated:

### Trapezoidal Rule

The trapezoidal rule approximates the area under a curve by dividing it into small trapezoids. It is simple and works well for evenly spaced environmental measurements.

### Simpson’s Rule

Simpson’s rule provides a more accurate approximation by using curved polynomial estimates between data points. It generally produces better results than the trapezoidal rule when the data changes smoothly.

```python
water_values = np.sin(x)+2


h = 1


trap_result = trapezoidal_rule(
    water_values,
    h
)


simpson_result = simpsons_rule(
    water_values,
    h
)


print("Trapezoidal Result:",trap_result)

print("Simpson Result:",simpson_result)
```

## Integration Method Comparison

The trapezoidal rule and Simpson's rule were compared to evaluate their
performance in estimating accumulated water quantities from sampled irrigation
data.

The trapezoidal rule approximates the area using straight-line segments between
data points, making it simple and computationally efficient. Simpson's rule uses
quadratic approximations and generally provides higher accuracy when the data
changes smoothly.

The comparison helps determine the most suitable numerical integration approach
for estimating cumulative irrigation requirements.

By solving these systems, the model can determine the optimal allocation of resources between different zones.

Gaussian elimination transforms a system of equations into a simpler form and calculates unknown values through a sequence of elimination and substitution steps.

```python
A = np.array([
    [2,1],
    [1,3]
])


b = np.array([
    10,
    15
])


solution = gaussian_elimination(
    A,
    b
)


print("Solution:")
print(solution)
```
